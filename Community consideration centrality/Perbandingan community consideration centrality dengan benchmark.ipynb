```python
import pandas as pd
import matplotlib.pyplot as plt
```

# All


```python
df=pd.read_excel('cc_dcb_c_plus_benchmark.xlsx') 
#df=pd.read_excel('cc_dcb_c_plus_benchmark distictif alpha 2.xlsx') 
df=df.head(50)
df=df.set_index('Nomor')
df.head(5)
```

```python
# https://www.statology.org/plot-multiple-lines-matplotlib/
# Jackknnife curve 
# An iteration model for identifying essential proteins by combining comprehensive PPI network with biological information
# United neighborhood closeness centrality and orthology for predicting essential proteins
#plt.figure(figsize=(16,12))

plt.figure(figsize=(16,12)) # Versi tipis

# plot lines satu per satu, dengan custom highlight
plt.plot(df['0.1 DC'], label='Community-consideration Degree Centrality, Î± = 0.1',linewidth=2, color='green')
plt.plot(df['0.2 DC'], label='Community-consideration Degree Centrality , Î± = 0.2', linewidth=2, color='blue' )
plt.plot(df['0.1 CC'], label='Community-consideration Closeness Centrality, Î± = 0.1', linewidth=1, color='red' )
plt.plot(df['0.2 CC'], label='Community-consideration Closeness Centrality, Î± = 0.2', linewidth=1, color='orange' )
plt.plot(df['0.1 BC'], label='Community-consideration Betweenness Centrality, Î± = 0.1', linewidth=1, color='purple' )
plt.plot(df['0.2 BC'], label='Community-consideration Betweenness Centrality, Î± = 0.2 ', linewidth=1, color='cyan' )
plt.plot(df['eigenvector_centrality'], label='eigenvector centrality', linewidth=3,  linestyle=(0, (1, 2)) )
plt.plot(df['katz_centrality'], label='katz centrality', linewidth=3,  linestyle=(0, (1, 2)) )
#plt.plot(df['current_flow_closeness_centrality'], label='current_flow_closeness_centrality', linewidth=3,  linestyle=(0, (1, 2)) )
plt.plot(df['information_centrality'], label='information centrality', linewidth=3,  linestyle=(0, (1, 2)) )
#plt.plot(df['current_flow_betweenness_centrality'], label='current_flow_betweenness_centrality', linewidth=1,  linestyle=(0, (1, 2)) )
plt.plot(df['communicability_betweenness_centrality'], label='communicability betweenness centrality', linewidth=3,  linestyle=(0, (1, 2)) )
plt.plot(df['load_centrality'], label='load centrality', linewidth=3,  linestyle=(0, (1, 2)) )
plt.plot(df['subgraph_centrality'], label='subgraph centrality', linewidth=3,  linestyle=(0, (1, 2)) )
plt.plot(df['subgraph_centrality_exp'], label='subgraph centrality exp', linewidth=3,  linestyle=(0, (1, 2)) )
plt.plot(df['harmonic_centrality'], label='harmonic centrality', linewidth=3,  linestyle=(0, (1, 2)) )
plt.plot(df['second_order_centrality'], label='second order centrality', linewidth=3,  linestyle=(0, (1, 2)) )
#plt.plot(df['brandes_betweenness_centrality'], label='brandes_betweenness_centrality', linewidth=1,  linestyle=(0, (1, 2)) )
plt.plot(df['D1'], label='distinctive centrality', linewidth=3,  linestyle=(0, (1, 2)) )
#plt.plot(df['D2'], label='D2', linewidth=5,  linestyle=(0, (1, 2)) )
#plt.plot(df['D3'], label='D3', linewidth=5,  linestyle=(0, (1, 2)) )
#plt.plot(df['D4'], label='D4', linewidth=5,  linestyle=(0, (1, 2)) )
#plt.plot(df['D5'], label='D5', linewidth=5,  linestyle=(0, (1, 2)) )
plt.plot(df['viral_rank_scores'], label='viral rank scores', linewidth=3,  linestyle=(0, (1, 2)) )
#plt.plot(df[''], label='', linewidth=1,  linestyle=(0, (1, 2)) )
#plt.plot(df[''], label='', linewidth=1,  linestyle=(0, (1, 2)) )
#plt.plot(df[''], label='', linewidth=1,  linestyle=(0, (1, 2)) )


#tambah legend
plt.legend(loc='upper left', title='Algoritma centrality')


#tambahkan  axis label
#plt.ylabel('Number of hallmark of cancer revealed \n on degree, closeness and betweenness centrality', fontsize=14)
#plt.xlabel('Number of top ranked proteins', fontsize=14)

#Versi bahasa Indonesia
plt.ylabel('Banyaknya $\it{hallmark\ of\ cancer}$ yang terungkap berdasar \n berbagai algoritma  $\it{centrality}$', fontsize=14)
plt.xlabel('Banyaknya proteins yang diperoleh dari berbagai algoritma  $\it{centrality}$', fontsize=14)

#plt.title('Titel di figure', fontsize=16)

#display plot
plt.show()
```
