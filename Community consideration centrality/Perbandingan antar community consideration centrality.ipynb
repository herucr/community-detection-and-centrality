```python
import pandas as pd
import matplotlib.pyplot as plt
```

# Degree Centrality

```python
df=pd.read_excel('cc_degree_c.xlsx') 
df=df.head(50)
df=df.set_index('Nomor')
df.head(5)
```

```python
# https://www.statology.org/plot-multiple-lines-matplotlib/
# Jackknnife curve 
# An iteration model for identifying essential proteins by combining comprehensive PPI network with biological information
# United neighborhood closeness centrality and orthology for predicting essential proteins
plt.figure(figsize=(16,12))

# plot lines satu per satu, dengan custom highlight
plt.plot(df[0], label='0.0',linewidth=1 , color='red')
plt.plot(df[0.1], label='0.1', linewidth=2, color='green' )
plt.plot(df[0.2], label='0.2', linewidth=2, color='orange' )
plt.plot(df[0.3], label='0.3', linewidth=0.5, color='purple', linestyle=(0, (1, 2)) )
plt.plot(df[0.4], label='0.4', linewidth=0.5, color='cyan',linestyle=(0, (1, 2)) )
plt.plot(df[0.5], label='0.5', linewidth=0.5, linestyle=(0, (1, 2)) )
plt.plot(df[0.6], label='0.6', linewidth=0.5, linestyle=(0, (1, 2)) )
plt.plot(df[0.7], label='0.7', linewidth=0.5, linestyle=(0, (1, 2)) )
plt.plot(df[0.8], label='0.8', linewidth=0.5, linestyle=(0, (1, 2)) )
plt.plot(df[0.9], label='0.9', linewidth=0.5, linestyle=(0, (1, 2)) )
plt.plot(df[1], label='1.0',linewidth=1 , color='blue')

#tambah legend
plt.legend(loc='upper left', title='alpha')


#tambahkan  axis label
plt.ylabel('Number of hallmark of cancer revealed on degree centrality', fontsize=14)
plt.xlabel('Number of top ranked proteins', fontsize=14)
#plt.title('Titel di figure', fontsize=16)

#display plot
plt.show()
```
